### Generating new categorical features

```python
import itertools

cat_features = [cname for cname in X.columns if
                    X[cname].nunique() < 10 and 
                    X[cname].dtype == "object"]

interactions = pd.DataFrame(index=X.index)
interactions_test = pd.DataFrame(index=X_test.index)

# Iterate through each pair of features, combine them into interaction features
for col1, col2 in itertools.combinations(cat_features, 2):
    new_col_name = '_'.join([col1, col2])

    # Convert to strings and combine
    new_values = X[col1] + "_" + X[col2]
    new_values_test = X_test[col1] + "_" + X_test[col2] # Test must have the same number of features

    interactions[new_col_name] = new_values  
    interactions_test[new_col_name] = new_values_test

#     print(interactions[new_col_name])
```
